```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from collections import Counter

import statsmodels.api as sm
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv("/content/data.csv") #reading the csv file and storing it as data
data = dataset.copy()
data.head(10)            #showing the dataset
```

```python
data = data.drop(['sex', 'cp',    'fbs',    'restecg',    'exang',    'slope',    'ca', 'thal', 'target'], axis = 1)
data_new = data.copy()

# Removing non-continuous features from the dataset
```

```python
# Calculating the distance between the data points
def distance_between(x1, x2):
  distance = (sum((x1 - x2)**2))**0.5
  return distance
```

```python
# Calculating the centroids
def findcentroids(clusters, datapoint):
    updated = []
    new_data = pd.concat([pd.DataFrame(datapoint),
                          pd.DataFrame(clusters,
                          columns=['K'])],axis=1)
    new_data.head(10)

# Using mean to update the centroid
    for i in set(new_data['K']):
        current = new_data[new_data['K'] == i][new_data.columns[:-1]]
        mean = current.mean(axis=0)
        updated.append(mean)
    
# Storing the cluster
    data['cluster_id'] = new_data['K']
    return updated
```

```python
# Finding the farthest centroids from the data points
def Farthest_Centroids(k, data):

    centroid = []
    
    for m in data:
        distance = []
        for n in k:
            distance.append(distance_between(m, n))

        # Appending the distance to later check the change in distance with each iteration
        # Calculating maximum distance between data points for the centroids
        centroid.append(np.argmax(distance))
    return centroid
```

```python
# Clustering solution using K-Means for a maximum of 100 iterations

def kmeans(data,k):

  X = np.array(data)

  # Choosing random centroids
  init_centroids = random.sample(range(0, len(data)), k)

  centroids = []
  for i in init_centroids:
      centroids.append(data.loc[i])
  centroids = np.array(centroids)
  get_centroids = Farthest_Centroids(centroids, X)
  

  for i in range(100):
        i+=1
        get_centroids = Farthest_Centroids(centroids, X)

        # Stop the loop if value stops changing
        if ((centroids[0] == findcentroids(get_centroids, X)[0]).all()):
          
          return centroids
          
        centroids = findcentroids(get_centroids, X)
        
  return centroids
```
